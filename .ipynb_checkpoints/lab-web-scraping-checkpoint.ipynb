```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/"
response = requests.get(url)
```

```python
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# The container for each book is an <article> tag with the class 'product_pod'
book_containers = soup.find_all('article', class_='product_pod')

# Create a list to hold all the full links
book_links = []
base_url = "http://books.toscrape.com/"

for book in book_containers:
    # Find the <a> tag inside the book's h3 tag
    a_tag = book.h3.a
    # Get the partial link from the href attribute
    partial_link = a_tag.get('href')
    # Create the full link and add it to our list
    full_link = base_url + partial_link
    book_links.append(full_link)

print(book_links[0]) # see the first full link
```

```python
first_book_url = book_links[0]

response = requests.get(first_book_url)
book_soup = BeautifulSoup(response.content, "html.parser")

print("Successfully fetched:", first_book_url)
```

```python
product_table = book_soup.find('table', class_='table-striped')

for row in product_table.find_all('tr'):
    header = row.find('th').get_text()
    if header == 'UPC':
        upc = row.find('td').get_text()
        print(f"Found UPC: {upc}")
    elif header == 'Availability':
        availability = row.find('td').get_text()
        print(f"Found Availability: {availability}")
```

```python
description_div = book_soup.find('div', id='product_description')

description_p = description_div.find_next_sibling('p')

description = description_p.get_text(strip = True)

print(description)
```

```python
# A list to store the dictionaries of book data
all_books_data = []

# A dictionary to convert star ratings from text to numbers
rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}

# Find all book containers on the first page
book_containers = soup.find_all('article', class_='product_pod')

# Loop through each book container
for book in book_containers:
    # Scrape data from the main page
    title = book.h3.a.get('title')
    price_str = book.find('p', class_='price_color').get_text(strip=True)
    rating_text = book.find('p', class_='star-rating').get('class')[1]

    # Visit the detail page for more info
    partial_link = book.h3.a.get('href')
    full_link = base_url + partial_link

    # Fetch and parse the detail page
    book_response = requests.get(full_link)
    book_soup = BeautifulSoup(book_response.content, "html.parser")

    # Scrape data from the detail page
    product_table = book_soup.find('table', class_='table-striped')
    upc = product_table.find('td').get_text() # UPC is the first <td> in the table
    availability = product_table.find_all('td')[5].get_text(strip=True) # Availability is the 6th <td>
    
    description_div = book_soup.find('div', id='product_description')
    description_p = description_div.find_next_sibling('p')
    description = description_p.get_text(strip=True) if description_p else "No description available"

    # --- Store the data in a dictionary ---
    book_data = {
        "UPC": upc,
        "Title": title,
        "Price (£)": float(price_str.replace("£", "")),
        "Rating": rating_map.get(rating_text, 0), # Use .get() for safety
        "Genre": "N/A", # We'll get this later
        "Availability": availability,
        "Description": description
    }

    all_books_data.append(book_data)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(all_books_data)
df.head()
```

```python
def scrape_books(min_rating, max_price):
    """
    Scrapes book data from books.toscrape.com across multiple pages,
    filters by minimum rating and maximum price, and returns a pandas DataFrame.
    """
    
    # A list to store the dictionaries of book data
    all_books_data = []
    
    # A dictionary to convert star ratings from text to numbers
    rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
    
    # Start with the first page
    current_url = "http://books.toscrape.com/catalogue/page-1.html"
    base_url = "http://books.toscrape.com/catalogue/"

    while current_url:
        print(f"Scraping page: {current_url}")
        response = requests.get(current_url)
        
        # Make sure the request was successful
        if response.status_code != 200:
            break
            
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Find all book containers on the page
        book_containers = soup.find_all('article', class_='product_pod')

        for book in book_containers:
            # Scrape rating and price first for filtering
            price_str = book.find('p', class_='price_color').get_text(strip=True)
            price = float(price_str.replace("£", ""))
            
            rating_text = book.find('p', class_='star-rating').get('class')[1]
            rating = rating_map.get(rating_text, 0)
            
            # --- FILTERING LOGIC ---
            # If the book meets the criteria, scrape the rest of the details
            if rating >= min_rating and price <= max_price:
                
                # --- Scrape remaining data ---
                title = book.h3.a.get('title')
                
                # Visit the detail page for more info
                partial_link = book.h3.a.get('href')
                full_link = base_url + partial_link
                
                book_response = requests.get(full_link)
                book_soup = BeautifulSoup(book_response.content, "html.parser")
                
                # Scrape data from the detail page
                upc = book_soup.find('th', string='UPC').find_next('td').get_text()
                availability = book_soup.find('th', string='Availability').find_next('td').get_text(strip=True)
                description_p = book_soup.find('div', id='product_description').find_next_sibling('p')
                description = description_p.get_text(strip=True) if description_p else "N/A"
                genre = book_soup.find('ul', class_='breadcrumb').find_all('li')[2].a.get_text(strip=True)

                # Store the data in a dictionary
                book_data = {
                    "UPC": upc,
                    "Title": title,
                    "Price (£)": price,
                    "Rating": rating,
                    "Genre": genre,
                    "Availability": availability,
                    "Description": description
                }
                all_books_data.append(book_data)

        # Find the 'next' button to go to the next page
        next_button = soup.find('li', class_='next')
        if next_button:
            next_page_partial_url = next_button.a.get('href')
            current_url = base_url + next_page_partial_url
        else:
            # If no 'next' button, we are on the last page
            current_url = None
            
    # Convert the list of dictionaries to a DataFrame
    return pd.DataFrame(all_books_data)

# --- Now, execute the function with your specified criteria ---
final_df = scrape_books(min_rating=4, max_price=20)

print("\\nScraping complete!")
final_df
```
